```python
#匯入credits(包含每個電影的演員名單、工作人員名單和電影id)
#!pip3 install pandas

import pandas as pd

credits = pd.read_csv(r"C:\Users\howger\Desktop\Movies_dataset\credits.csv",sep=',' )
credits
```

```python
#分析每個電影的演員和工作人員人數
import ast

credits_count = pd.DataFrame({
    'cast_count':credits['cast'].apply(ast.literal_eval).apply(len),
    'crew_count':credits['crew'].apply(ast.literal_eval).apply(len),
    'id':credits['id']})

credits_count.head()
```

```python
#探索資料
credits_count.describe()
```

```python
#載入主要電影資訊
movies_info = pd.read_csv(r"C:\Users\howger\Desktop\Movies_dataset\movies_metadata.csv",sep=',' )
movies_info['original_title'] = movies_info['original_title'].fillna('').astype(str)
movies_info
```

```python
movies_info = movies_info[pd.to_numeric(movies_info['id'], errors='coerce').notna()] #將無法轉成數值的值(如日期)刪除
movies_info['id'] = movies_info['id'].astype(int)
```

```python
credits_count = pd.merge(credits_count, movies_info, how='left', on='id')[['cast_count','crew_count','original_title','id']]
credits_count
```

```python
max_castindex , max_crewindex = credits_count["cast_count"].idxmax() , credits_count["crew_count"].idxmax()
max_movienamecast , max_movienamecrew = credits_count.loc[max_castindex,'original_title'] , credits_count.loc[max_crewindex,'original_title']
max_castcount , max_crewcount = credits_count['cast_count'].max() , credits_count['crew_count'].max()

print(f'最多演員的電影id為:{ max_castindex },電影名稱為:{max_movienamecast},演員數量為：{ max_castcount }')
print(f'最多工作人員的電影id為:{ max_crewindex },電影名稱為:{max_movienamecrew},演員數量為：{ max_crewcount }')
```

```python
ratings = pd.read_csv(r"C:\Users\howger\Desktop\Movies_dataset\ratings_small.csv",sep = ',')
ratings
```

```python
keywords['keywords_text'] = keywords['keywords'].apply(ast.literal_eval).apply(lambda x:' '.join([a['name'] for a in x]))
keywords
```

```python
#詞語出現次數的稀疏矩陣
#!pip3 install scikit-learn

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english',max_features=1000)
X = vectorizer.fit_transform(keywords['keywords_text'])

df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

print(df)
```

```python
#TF-IDF稀疏矩陣
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english',max_features=1000)
tfidf_matrix = vectorizer.fit_transform(keywords['keywords_text'])


df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
print(df_tfidf)
```

```python
#做每一個電影(id以及名稱)的關鍵詞出現頻率最高的表
import numpy as np

tfidf_dense_matrix = tfidf_matrix.toarray() 

top_tfidf_indices = np.argmax(tfidf_dense_matrix, axis=1)

top_tfidf_words = [vectorizer.get_feature_names_out()[i] for i in top_tfidf_indices]

top_tfidf_values = np.max(tfidf_dense_matrix, axis=1)

df = pd.DataFrame({
    'Movie_id': keywords['id'] ,
    'Top Word': top_tfidf_words,
    'TF-IDF Value': top_tfidf_values
})

merged_df = pd.merge(df, movies_info[['id', 'original_title']], left_on='Movie_id', right_on='id', how='left') #串結兩資料集的id找出電影名稱

print(merged_df[['Movie_id', 'original_title', 'Top Word', 'TF-IDF Value']]) 

```

```python
# 過濾出 TF-IDF 值為 1 的資料
tfidf_1_df = merged_df[merged_df['TF-IDF Value'] == 1]

# 使用 Movie_id 從 movies_info 資料集取出對應的電影名稱
result_1_df = pd.merge(tfidf_1_df, movies_info[['id', 'original_title']], how='left', left_on='Movie_id', right_on='id')

# 顯示結果
print(result_1_df[['original_title_x','Top Word']])
```

```python
#畫出TF-IDF為1的電影關鍵詞的圖
#!pip install seaborn
import matplotlib.pyplot as plt

category_counts = result_1_df['Top Word'].value_counts()

plt.figure(figsize=(6,6))  # 設置圖形大小
category_counts[category_counts >= 100].plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99'])
plt.title('TOP 5 Movies Keywords')  # 圖表標題
plt.ylabel('')  # 去除y軸標籤
plt.show()

###由圖可知，較特殊且出現超過100次的電影關鍵詞有musical、sport以及biography、suspese等等...
```

```python
#查資料庫裡的電影名稱

def search_movies_names(y):
    movies_names = movies_info[movies_info['original_title'].str.contains(y, case=False, na=False)][['id','original_title']]

    if movies_names.empty:
        return "沒有找到匹配的電影"
    
    return movies_names

search_movies_names("Star War")
    
```

```python
#方法一、利用TF-IDF計算cos相似度矩陣並取出欲推薦的前五個電影

from sklearn.metrics.pairwise import cosine_similarity

def get_top_k_similar_movies(keywords, k=5):

    filtered_movies = movies_info[movies_info['original_title'].str.contains(keywords, case=False, na=False)]

    if filtered_movies.empty:
        return "沒有找到匹配的電影"
    
    movie_idx = filtered_movies.iloc[0]['id']
    
    search_movie_name = filtered_movies.iloc[0]['original_title']

    keywords_id = movies_info[movies_info['id'] == movie_idx].index[0] #將movie_info的id轉換成keywords的列指標

    #計算cos相似性矩陣
    similarities = cosine_similarity(tfidf_matrix[keywords_id], tfidf_matrix).flatten() #如果需要取相關電影推薦，只計算相關的電影相似度來節省記憶體空間

    top_k_indices = np.argsort(similarities)[::-1][1:k+1]  
    top_k_indices_corrid = movies_info.iloc[top_k_indices]['id'] #再將找出來的前5個keywords列指標轉換回來movie_info的id
    top_k_indices_dataframe = pd.DataFrame({'id':top_k_indices_corrid})

    #找movie_info前5個推薦電影的id和original_title並改欄位名稱
    top_k_movies_name = pd.merge( top_k_indices_dataframe , movies_info[['id','original_title']], how='left' ).rename(columns={'id':'電影id','original_title':'電影名稱'}) 

    scores = pd.DataFrame({'Cos相似度分數':similarities[top_k_indices]})
    result = pd.concat([top_k_movies_name,scores],axis=1)
    search_movie_name = movies_info.loc[movies_info['id'] == movie_idx,'original_title'].values[0]
    return movie_idx,search_movie_name,result 



movie_idx,search_info,top_movies_name  = get_top_k_similar_movies("Star")

print("你搜尋的電影名稱為：\n",movie_idx,search_info,"\n")
print("Top 5 推薦電影：\n",top_movies_name)
```
